## Setup

```python
from polis_client import PolisClient
from pprint import pprint

polis = PolisClient()
```

## Interacting with the default Pol.is server

```python
comments = polis.get_comments(conversation_id="2demo")
if comments:
    pprint(comments[0].to_dict())
```

```python
report = polis.get_report(report_id="r68fknmmmyhdpi3sh4ctc")
if report:
    pprint(report.to_dict())
```

```python
conversation = polis.get_conversation(conversation_id="2demo")
if conversation:
    pprint(conversation.to_dict())
```

```python
math = polis.get_math(conversation_id="2demo")
pprint(math.to_dict(), depth=1)
```

```python
votes = polis.get_votes(conversation_id="2demo", pid=2)
if votes:
    pprint(votes[0].to_dict())
```

## Using an xid

```python
polis_auth = PolisClient(xid="foobar")
vote = polis_auth.create_vote(conversation_id="2demo", tid=2, vote=1)

reusable_jwt_token = polis_auth._token

polis_auth = PolisClient(token=reusable_jwt_token)
vote = polis_auth.create_vote(conversation_id="2demo", tid=2, vote=0)

mypid = vote.to_dict()["currentPid"]

votes = polis_auth.get_votes(conversation_id="2demo", pid=mypid)
if votes:
    votes = [v.to_dict() for v in votes]
    pprint([v for v in votes if v["tid"] == 2])

# Update our token if it was refreshed internally.
if reusable_jwt_token != polis_auth._token:
    reusable_jwt_token = polis_auth._token
```
