# Gallery Example: MAP/M/k Queue

This example demonstrates a MAP/M/k queueing system:
- **Arrivals**: Markovian Arrival Process (MAP)
- **Service**: Exponential service times
- **Servers**: k servers (default k=2)
- **Capacity**: Infinite
- **Scheduling**: FCFS

This extends the MAP/M/1 queue to multiple servers.

```python
from line_solver import *
GlobalConstants.set_verbose(VerboseLevel.STD)
```

```python
def gallery_mapmk(map_dist=None, k=2):
    """Create MAP/M/k queueing model"""
    if map_dist is None:
        map_dist = MAP.rand()
    
    model = Network('MAP/M/k')
    
    # Block 1: nodes
    source = Source(model, 'mySource')
    queue = Queue(model, 'myQueue', SchedStrategy.FCFS)
    sink = Sink(model, 'mySink')
    
    # Block 2: classes
    oclass = OpenClass(model, 'myClass')
    source.set_arrival(oclass, map_dist)
    queue.set_service(oclass, Exp(2))
    queue.set_number_of_servers(k)
    
    # Block 3: topology
    model.link(Network.serial_routing(source, queue, sink))
    
    return model

# Create the model with a random MAP and k=2 servers
model = gallery_mapmk(k=2)
```

## About MAP/M/k Queues

The MAP/M/k queue combines:
- **Correlated arrivals** (via MAP): Can model bursty, self-similar, or autocorrelated traffic
- **Multiple servers**: Provides parallelism and reduces waiting times
- **Exponential service**: Memoryless service times

With k servers, the maximum service capacity is k × μ where μ is the service rate per server.

```python
# Solve with multiple solvers
print("\n=== Solver Results ===")

# CTMC Solver
solver_ctmc = CTMC(model, cutoff=15)
avg_table_ctmc = solver_ctmc.avg_table()
print("\nCTMC Solver:")
print(avg_table_ctmc)

# Fluid Solver
solver_fluid = FLD(model)
avg_table_fluid = solver_fluid.avg_table()
print("\nFluid Solver:")
print(avg_table_fluid)
```
