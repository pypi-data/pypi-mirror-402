# Gallery Example: U/M/1 Queue

This example demonstrates a U/M/1 queueing system:
- **Arrivals**: Uniform distribution
- **Service**: Exponential service times
- **Servers**: 1 server
- **Capacity**: Infinite
- **Scheduling**: FCFS

The uniform distribution provides bounded inter-arrival times with moderate variability.

```python
from line_solver import *
GlobalConstants.set_verbose(VerboseLevel.STD)
```

```python
def gallery_um1():
    """Create U/M/1 queueing model"""
    model = Network('U/M/1')
    
    # Block 1: nodes
    source = Source(model, 'Source')
    queue = Queue(model, 'Queue', SchedStrategy.FCFS)
    sink = Sink(model, 'Sink')
    
    # Block 2: classes
    oclass1 = OpenClass(model, 'Class1')
    # Uniform inter-arrival time between 1 and 2 (mean = 1.5)
    source.set_arrival(oclass1, Uniform(1, 2))
    queue.set_service(oclass1, Exp(2))
    
    # Block 3: topology
    P = model.init_routing_matrix()
    P.add_route(oclass1, source, queue, 1.0)
    P.add_route(oclass1, queue, sink, 1.0)
    model.link(P)
    
    return model

# Create the model
model = gallery_um1()
```

## About Uniform Distribution

For Uniform(1, 2) inter-arrival times:
- **Mean**: (1 + 2) / 2 = 1.5
- **Variance**: (2 - 1)² / 12 = 1/12 ≈ 0.083
- **SCV (squared coefficient of variation)**: 1/12 / 1.5² = 1/27 ≈ 0.037

The low SCV means arrivals are more regular than Poisson (SCV = 1).

```python
# Solve with multiple solvers
print("\n=== Solver Results ===")

# CTMC Solver
solver_ctmc = CTMC(model, cutoff=15)
avg_table_ctmc = solver_ctmc.avg_table()
print("\nCTMC Solver:")
print(avg_table_ctmc)

# Fluid Solver
solver_fluid = FLD(model)
avg_table_fluid = solver_fluid.avg_table()
print("\nFluid Solver:")
print(avg_table_fluid)
```
