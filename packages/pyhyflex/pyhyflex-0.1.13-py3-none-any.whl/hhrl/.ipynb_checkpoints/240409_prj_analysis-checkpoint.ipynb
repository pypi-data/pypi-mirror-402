```python
import os

def get_file_contents(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()
    except UnicodeDecodeError:
            return "<Unable to read file contents due to encoding issue>"

def generate_project_summary(root_dir):
    project_summary = ""
    for root, dirs, files in os.walk(root_dir):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            project_summary += f"Title: {file_name}\n"
            project_summary += f"Location: {file_path}\n"
            project_summary += "Content:\n"
            project_summary += get_file_contents(file_path)
            project_summary += "\n\n"
    return project_summary

def save_summary_to_file(summary, output_file):
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(summary)

if __name__ == "__main__":
    root_directory = '.'  # 当前文件夹
    summary = generate_project_summary(root_directory)
    output_file = 'hh_project_summary.txt'  # 输出文件名
    save_summary_to_file(summary, output_file)
    print("Summary generated and saved to", output_file)
```
