#### April10 并行线程同时运行多个实例 

```python
import subprocess
from concurrent.futures import ThreadPoolExecutor
import random
import time
from datetime import datetime



# 定义问题列表
_problems = ["TSP", "FS", "SAT", "BP", "VRP", "PS"]

# 定义固定的参数
_agent = "DQN"  # 可选项: ["DQN", "DQNUCB", "DMAB", "FRRMAB", "RAND", "QL"]
_state = "S7"   # 可选项: ["S7"]
_reward = "RIP" # 可选项: ["RIP", "IR", "DIV", "IND", "IOD", "IOP", "DIP"]
_acceptance = "ALL"
#instance 0-9
_round = random.randint(1, 31) # 可选项: {1..31}
_time = 3000


# 定义脚本执行路径和输出路径
_python = "/home/Chaofan_Tu/miniconda3/bin/conda run -n hyflex python"
_config = "/home/Chaofan_Tu/Documents/hyflex3/hhrl/configs/fir_discrete.ini"
_output_dir = "/home/Chaofan_Tu/Documents/hyflex3/hhrl/results"
_max_workers = 3

# 定义运行命令的函数
def run_command(command, task_id):
    start_time = time.time()  # 记录开始时间
    # 获取当前时间和将当前时间格式化为字符串
    print(f"Task {task_id} command: {command}")  # 打印子任务id和具体命令内容
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    out, err = process.communicate()
    print(f"Task {task_id} output: {out}")  # 打印子任务
    print(f"Task {task_id} error: {err}")  # 打印子任务
    end_time = time.time()  # 记录结束时间
    print("Current time:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    print(f"Task {task_id} completed in {end_time - start_time:.2f} seconds.")  # 打印子任务id和完成时间
    print("=" * 50)

# 主程序入口
if __name__ == "__main__":
    start_time = time.time()
    print("Current time:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    print("Starting tasks...")
    # 定义要执行的命令列表
    commands = []
    for task_id, _problem in enumerate(_problems):
        command = f'{_python} runner.py -p {_problem} -c {_config} -ag {_agent} -st {_state} -rw {_reward} -ac {_acceptance} -r {_round} -t {_time} -o {_output_dir}'
        commands.append((command, task_id))

    # 使用 ThreadPoolExecutor 创建线程池，限定最大线程数为 max_workers
    with ThreadPoolExecutor(max_workers=_max_workers) as executor:
        # 提交任务给线程池并获取 Future 对象列表
        futures = [executor.submit(run_command, command, task_id) for command, task_id in commands]
        
        # 打印子进程的启动信息
        print("Tasks started.")

        # 等待所有任务完成
        results = []
        for future_id, future in enumerate(futures):
            print(f"Result {future_id} : {future.result()} ")

    end_time = time.time()
    print("Current time:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    print(f"All tasks completed. Total time: {end_time - start_time:.2f} seconds.")
```
