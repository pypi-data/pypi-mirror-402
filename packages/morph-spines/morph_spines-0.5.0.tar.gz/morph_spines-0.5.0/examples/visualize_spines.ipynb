# Visualization example: neuron section spines

```python
# Imports
from trimesh import util as triutil

from morph_spines.utils.morph_spine_loader import load_spines
```

```python
# Load spine data from morphology-with-spines H5 file
filepath = "./data/morphology_with_spines/864691134884740346.h5"
spines = load_spines(filepath)
```

```python
# Get the spines mesh for a given neuron's section ID and plot them
section_id = 9
mesh = triutil.concatenate(spines.spine_meshes_for_section(section_id))
mesh.show()
```
