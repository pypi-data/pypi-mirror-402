```python
import sqlite3

# Connect to your database
conn = sqlite3.connect("/tmp/vllmonitor/models.db")
cursor = conn.cursor()

# List all tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables:", tables)

# Show table schema
cursor.execute("PRAGMA table_info(table_row);")
schema = cursor.fetchall()
print("Schema for table_row:")
for col in schema:
    print(col)

# Preview data
cursor.execute("SELECT * FROM table_row LIMIT 5;")
rows = cursor.fetchall()
print("First rows:")
for row in rows:
    print(row)

conn.close()
```
