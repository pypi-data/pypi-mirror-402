```python
import os

from rich import print

from myosdk import Client

key = os.getenv(
    "MYOSDK_API_KEY", "v2m_KZD5F99K5wMkJR3z:7QXducgg3wapXv9Fla2FdQAcKMqJfZXK"
)

client = Client(api_key=key, base_url="http://localhost:8000")

```

```python
print(client.jobs.list(status=["SUCCEEDED", "FAILED"], limit=20))
```

```python
# Upload video
input_asset = client.assets.upload_file(
    "/home/ouassim/dev/myoGalleryKinesis/tmp-assets/input.mp4"
)
print(input_asset)
```

```python
# Start a tracker job
job = client.jobs.start_tracker(
    video_asset_id=input_asset["asset_id"],
    params={"fps": 30, "skeleton": "humanoid"}
)

# Wait for job to complete
result = client.jobs.wait(job["job_id"])
input_asset_id = result["output"]["trackers_asset_id"]
print(input_asset_id)
```

```python
# Get the tracker asset
input_asset = client.assets.get(input_asset_id)
print(input_asset)

# Download the tracker asset
client.assets.download(input_asset_id, "/home/ouassim/dev/myoGalleryKinesis/tmp-assets/output.pkl")
```

```python
# Upload tracker pickle file
trk_asset = client.assets.upload_file(
    "/home/ouassim/dev/myoGalleryKinesis/tmp-assets/output.pkl"
)
print(trk_asset)
```
