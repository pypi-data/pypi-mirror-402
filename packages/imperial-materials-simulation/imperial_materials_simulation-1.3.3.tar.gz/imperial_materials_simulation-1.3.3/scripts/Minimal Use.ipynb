```python
import imperial_materials_simulation as ims
```

```python
sim = ims.Simulation(n_atoms=21, starting_temperature=800, microstructure=ims.starting_microstructure)
sim.display(live_display_interval=5_000) #updates live as simulations are run
```

```python
sim.NVT_run(n_steps=100_000, temperature=1200, gamma=0.005, integrator='OVRVO')
sim.relax_run(n_steps=100_000, step_size=0.03)
sim.MMC_run(n_steps=500_000, temperature=800, random_scale=0.1)
sim.NVE_run(n_steps=100_000, temperature=600, gamma=0.005, integrator='Verlet')
```

```python
sim.to_file("simulation object.ims")
sim.to_excel("simulation data.xlsx")
```
